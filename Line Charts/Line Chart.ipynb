```python
# Import the base libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Load the dataset
df_spx_ftse_00_10 = pd.read_csv("C:\\Users\\User\\Desktop\\data_science bootcamp\\Dataset\\D.V.Datasets\\line_chart_data.csv")
```

```python
df_spx_ftse_00_10
```

```python
# Convert the "Date" Values to "date-time" format
df_spx_ftse_00_10["new_date"] = pd.to_datetime(df_spx_ftse_00_10["Date"])
```

```python
df_spx_ftse_00_10["new_date"]
```

```python
labels = ["S&P 500"], ["FTSE 100"]
# Increase the size of the plot
plt.figure(figsize = (20, 8))
# Plot the graph
plt.plot(df_spx_ftse_00_10["new_date"], df_spx_ftse_00_10["GSPC500"])
plt.plot(df_spx_ftse_00_10["new_date"], df_spx_ftse_00_10["FTSE100"])
# Label the chart
plt.title("S&P VS FTSE Returns (2000 - 2010)", fontsize = 14, fontweight = "bold") 
plt.ylabel("Returns")
plt.xlabel("Date")
# Legend for clarity
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
# Chose a smaller timeframe / slice the data
df_spx_ftse_00_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10.new_date >= '2008-07-01') &
                                         (df_spx_ftse_00_10.new_date <= '2008-12-31')]
```

```python
df_spx_ftse_00_H2_08
```

```python
# Labels = ["S&P 500"], ["FTSE 100"]
# Increase the size of the plot
plt.figure(figsize = (20, 8))
# Plot the graph
plt.plot(df_spx_ftse_00_H2_08["new_date"], df_spx_ftse_00_H2_08["GSPC500"], color = "midnightblue")
plt.plot(df_spx_ftse_00_H2_08["new_date"], df_spx_ftse_00_H2_08["FTSE100"], color = "crimson")
# Label the chart
plt.title("S&P VS FTSE Returns (H2 2008)", fontsize = 14, fontweight = "bold") 
plt.ylabel("Returns")
plt.xlabel("Date")
# Legend for clarity
plt.legend(labels = labels, fontsize = "large")
plt.show()
```
